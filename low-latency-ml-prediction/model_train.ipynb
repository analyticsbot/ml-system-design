```python
import mlflow
import mlflow.sklearn
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score
from datetime import datetime

# Set MLflow tracking URI
mlflow.set_tracking_uri("http://localhost:5002")  # Point to the mounted volume

# Define run name with prefix and current datetime
run_name = f"MyExperiment_{datetime.now().strftime('%Y%m%d_%H%M%S')}"

# Load data
data = load_iris()
X_train, X_test, y_train, y_test = train_test_split(data.data, data.target, test_size=0.3, random_state=42)

# Start an MLflow run
with mlflow.start_run(run_name=run_name):
    # Train a model
    model = RandomForestClassifier(n_estimators=100, random_state=42)
    model.fit(X_train, y_train)

    # Predict and calculate accuracy
    predictions = model.predict(X_test)
    accuracy = accuracy_score(y_test, predictions)

    # Log model, metrics, and parameters
    mlflow.log_param("n_estimators", 100)
    mlflow.log_metric("accuracy", accuracy)
    mlflow.sklearn.log_model(sk_model=model, artifact_path="model")  # Specify artifact location

    # Register the model
    model_name = "IrisClassifier"
    mlflow.register_model("runs:/{}/model".format(mlflow.active_run().info.run_id), model_name)

print(f"Model trained and registered with accuracy: {accuracy}")
```
